```python
# ╔══════════════════════════════════════════════════════════════╗
# ║  🔓  UNLOCK YOUR EXAM  ·  IN221  ·  Eng. Ahmed Métwalli     ║
# ║  Enter your registration number and the exam key below.     ║
# ╚══════════════════════════════════════════════════════════════╝
# ⚠️  Do NOT modify this cell.

import hashlib, zlib, base64, json, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown, HTML, Javascript
from sklearn.compose import ColumnTransformer
from sklearn.exceptions import ConvergenceWarning
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression, Ridge, Lasso, ElasticNet, LinearRegression
from sklearn.metrics import (
    accuracy_score, f1_score, classification_report, confusion_matrix,
    ConfusionMatrixDisplay, r2_score, mean_absolute_error, mean_squared_error
)
from sklearn.model_selection import (
    train_test_split, KFold, StratifiedKFold, cross_validate,
    cross_val_score, GridSearchCV
)
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler, RobustScaler, MinMaxScaler

warnings.filterwarnings('ignore')
warnings.filterwarnings('ignore', category=ConvergenceWarning)

# ── Encrypted payloads (do not modify) ──────────────────────────
_VERIFY_TOKEN     = 'dRFV5HUVBIpurNMaZYF/XY/Dvi/6fwM8jlA='
_QUESTION_PAYLOAD = 'dRGLiaGLFLMM5vxruDaRGQj9LfPNC9ZZ/RFqf1TytQ45SApmmikGr4DAOwpEAmBvWWk7B4uXbH5wUtO3Bnk/451RaVg0ZFZNKsUz0hdFdnJh91SlxEYLRswl0pTGyDdb8yQYewNORtCmXl2rtVztULi9I3SwL25zQ1deTZ089f/eRBUkeBrv3ZbcLBP3iuS9LRIZl7J4m0UONcWSA2lzrpPAX8O1sVUeJpXZy5vdrhHthpchdY28D+PY5O62Zm9+cQuxRCTo9FPueS2SAXpdoS/cRuLkYH34OpZHYUUIWrJfl3lm/8uL1mv7AwnFVxByCaEEZ3S0fEXsxkny6rW93DGDwlue8mtrJ643aZHZ4ta8A2Stx21e6PfQcGOgUCpsRVRMvS1+GsOrHJX4edD4c2YYcTuTGCKIbBJPPQuMJlDXOY2TOb6kjsKahSOUzdTIJTFacMyqHQ8OfQ8Zlbzd7QdKCelZTvRms1/TFqxvUy+InkwNp0PbqJb5FF504mRrsGdgPt+PRm7t0ToYIkJOi2jsijA4J8B8tswc+eThmxGh4Nuo5pgWSWB6KzbQa0HjiOkBh+p0VtZKiVWfpqF+suRSztZFFKsMQ6lzkN6L18UeR+7f3OE7mZxKwVTlcZ2wqXZXEfdyJMeCVXPT9Gl2dwlrL1ob7LLcSDB3bzzPdtEDAwmMId0bVXzrRLqzypm0NxIhQmWC/6vFzUiaPEpAlwr3Vq92kRdQAO7G5mqOp0FVjCJ/aXi4KZ2BAuaPr+eq/cKEPlPxEbqp3DEOROvKD3ooqMO9xi4q+dwiy79CAS8yT7vhJuLCHx2ttfUHPiNOk9cOlN9oqOwdEr7OSeVF7XgWD91AaXXd8V/iuwxe69zK0OabeEMqwVnodF0hpCS8zyywkVtK2jY6hNZgIkZpUl/Sa3enz69dDtSEBIYr0/3A3WuS6ZLJcFWvrGDdYmhJS56wX6UFT9FZCc/cEn48I4Rqr+sC6aUl/uB31ToBNyxRxam0NyusfCPALlp697MHDf6wpruvz+iqdO/W8SxWay0v7liktGIqUNMXOHVpH6vu9zvl7W2/uQYBQtSpc7nOyVVRP+h5ACQd3Awh4wKyUUJ9b1e6Xz2tPNo+4wGUNm1e2ku3yImSDv6hbPBEiTXDZG53X92e/fLa1FKymwg9Dvba+aLQQME/a93EcJo1LMI5EiKZhT3+6TPDVhT2I00eU9KcdmDjh1evDsoQYNK6aGmdCkQo+8kRytjD6OblizTrxYN2e0aZjk+VH0k9cMqfnaXqdrsKZghOoRTzmpWE5kohOiYDPa3nWpNwFkd7c8DSBiXvIu13LX6AKewLTcTXXMt5pcWu9SM54X9XtVjBdncHj8sbBRqYhdfEkXQbGILZ24EAYSoBnbRAOUxDTuuKxNk4zNazeSB7WGY3pOKzAJ2cRkktMOTWSDiGduGgsLUqOoJcubGd5Vm/LNLRJNVGRGZGrgJo10LZ678qwIZonCXPTFcRPMk43mNva5X4rhPwvh9Bh74YqQA/jQVPC00B/tWuYn8XfU1aqVLvCl5TYD6pUTrLbDQ63x8Lr5qvavMaIl5HudZQRnU99X5crIYJUKspUkR6ukoD+GLtpL7pMr9VrnT8hNJRLlPrkJvbqkVX067zLyDY6n9mHwkD+zIom1V79oGFmKa3SJLM8dpsFAsbLfOvkQcrErwhWTsY1Dvqf5aqAGvWmE4vMYQeDAkYMdYV8Qui5Skpt/BExYlxomlawQMlfLIi3qTYzlCV7Qn6R6wPh3QxXGFtD3w1JPe8akqTFOKNE7gmJ9tIYuXUQ6HVMvVo2+D4mWMDKDmSBlj4awI0QIM9E4kYZCDLLGV2ji8KtYmzXIqPYImBoNSTC8CASU7z96k11WReOkuF7QgtzYa9VmvV3JVTgkby7KVnXVURF9yEzOeqyGiJp3S3o5PsXGeiGLoZrRCxOZvRGzOdwfQ717Ms0T1OXeBVYUB/wZEVpG5mCnufK/8s8F3W1cGJ1EZ7xXchOU4t/MlIcIRYIlY9tO7AU36THTKQBajJEupdOinbzLH+evOCrB3wqQHDvW1VaRm5ZNotuHHQ9TCMN/kA3yACpnAmGJL1Z6C9VdUCEVMbxevUhSZNZoa6+RUUKUmI3Z1c/S0a9bXGQP+VnTw24m51B4bflbl20vNnID1y0pXjOSwImNaG59ykkr19EiTXBicGuw5MMcZcIwZzdzlpCkKNHNne1DICXHQDo+AX3cUYQit1gC7Xl1j6EgLz/JkNjPymrpmXAp2wH/3qPpWJ8rEyNpX/tNDYNZ5rW/arU8bLzxSRnLpU/+aYgxrtwRZm+CPe6clhSFLiTmzD72mKp8r2bPIdvm8eZO60/ufhT/TS5T3kpQsEMBuwgnOYnQ0ZHVqGEmfOrHgszTvS8TeUmcSAJ3aA/v6VXS/8fKgYXmA7OiEX5P34ZzG3IBnoAq/lHWnZYN0BAPNZ75f0PM2lJ+w9z3uSXCAotwPVSDiaeR+aKz+nGAISVmXedP6/UIYeAX0I7A5C85waZcFc6e/qo+wGy1rHQJWthFNPbjHvYVyWS1y+GoBNmnHEOM9T/FA8vMs/WTSY9sORj9DIKmpUhoyLAVdOgbchvqSXUIg0zffBj/B738gOTzN+vSTFO/1GoMOiiBbXowm/E1uouGi8bDUTZyRQwGGPRaFNsuEaw7DdWMhIC5xaYAKSNWH/qUgj43eWF90uefmnE/SjiGw+bIUQ1aHESLWJgRLvzX1SI/vBsQiyXWX6f5IjLWT30sID5NLWTYE='

# ── Crypto helpers ───────────────────────────────────────────────
def _xor(data, pwd):
    k = hashlib.sha256(pwd.encode()).digest()
    return bytes([b ^ k[i % len(k)] for i, b in enumerate(data)])


def _unlock(encoded, pwd):
    try:
        return zlib.decompress(_xor(base64.b64decode(encoded), pwd)).decode('utf-8')
    except Exception:
        return None

# -- Dataset generator: Solar microgrid reliability ------------------------
def _make_dataset(seed):
    rng = np.random.default_rng(seed)
    n = 560

    regions = np.array(['Coastal', 'Desert', 'Delta', 'Upper Valley', 'Metro'])
    panel_types = np.array(['Mono PERC', 'Thin Film', 'Polycrystalline', 'Bifacial'])
    inverter_models = np.array(['HelioMax', 'VoltEdge', 'SunCore', 'GridAxis'])

    region_effect = {'Coastal': -0.5, 'Desert': 1.8, 'Delta': 0.2, 'Upper Valley': 1.0, 'Metro': -0.8}
    panel_effect = {'Mono PERC': 2.0, 'Thin Film': -1.3, 'Polycrystalline': -0.4, 'Bifacial': 2.8}
    inverter_effect = {'HelioMax': 0.8, 'VoltEdge': -0.4, 'SunCore': 0.2, 'GridAxis': -0.1}

    region = rng.choice(regions, n, p=[0.18, 0.25, 0.22, 0.20, 0.15])
    panel_type = rng.choice(panel_types, n, p=[0.35, 0.18, 0.27, 0.20])
    inverter_model = rng.choice(inverter_models, n, p=[0.30, 0.24, 0.26, 0.20])

    re = np.array([region_effect[x] for x in region])
    pe = np.array([panel_effect[x] for x in panel_type])
    ie = np.array([inverter_effect[x] for x in inverter_model])

    measurement_month = rng.integers(1, 13, n)
    seasonal = np.sin((measurement_month - 3) / 12 * 2 * np.pi)
    system_capacity_kw = np.clip(rng.normal(92, 18, n) + pe * 2 + re * 3, 35, 155)
    solar_irradiance = np.clip(rng.normal(5.4, 0.9, n) + re * 0.35 + seasonal * 0.75, 2.2, 8.4)
    cloud_cover_pct = np.clip(rng.normal(32, 16, n) - re * 5 - seasonal * 8, 0, 95)
    ambient_temp_c = np.clip(rng.normal(27, 5.5, n) + re * 1.3 + seasonal * 6, 8, 48)
    panel_temp_c = np.clip(
        ambient_temp_c + 7 + 1.7 * solar_irradiance - 0.035 * cloud_cover_pct + rng.normal(0, 2.2, n),
        15, 67
    )
    humidity_pct = np.clip(rng.normal(55, 18, n) - re * 6 + 0.26 * cloud_cover_pct, 10, 96)
    wind_speed_m_s = np.clip(rng.normal(4.6, 1.5, n) + re * 0.25, 0.3, 12)
    dust_index = np.clip(
        rng.gamma(2.3, 4.5, n) + np.maximum(re, 0) * 1.4 - 0.10 * humidity_pct + rng.normal(0, 1, n),
        0.5, 35
    )
    panel_age_months = rng.integers(2, 96, n)
    days_since_maintenance = np.clip(
        rng.normal(55, 28, n) + dust_index * 1.2 + panel_age_months * 0.15,
        1, 190
    )
    inverter_load_pct = np.clip(
        42 + 5.4 * solar_irradiance + 0.18 * system_capacity_kw - 0.17 * cloud_cover_pct
        + ie * 1.2 + rng.normal(0, 6, n),
        20, 99
    )
    voltage_stability_score = np.clip(
        98 - 0.21 * inverter_load_pct - 0.11 * panel_temp_c - 0.10 * dust_index
        + ie * 1.8 + rng.normal(0, 2.5, n),
        55, 99
    )
    battery_state_pct = np.clip(
        52 + 5.0 * solar_irradiance - 0.20 * cloud_cover_pct - 0.12 * inverter_load_pct
        + rng.normal(0, 8, n),
        5, 100
    )
    cleaning_cost_usd = np.clip(
        80 + 2.7 * dust_index + 0.55 * system_capacity_kw + 0.35 * days_since_maintenance + rng.normal(0, 14, n),
        60, 380
    )
    daily_energy_kwh = np.clip(
        22 + system_capacity_kw * (
            3.15 + 0.50 * solar_irradiance - 0.010 * cloud_cover_pct
            - 0.012 * dust_index - 0.004 * np.maximum(panel_temp_c - 35, 0)
        )
        + pe * 7 + ie * 4 - 0.28 * panel_age_months + rng.normal(0, 35, n),
        70, 790
    )
    fault_score = (
        -4.3 + 0.055 * dust_index + 0.020 * days_since_maintenance
        + 0.038 * (panel_temp_c - 35) + 0.042 * inverter_load_pct
        - 0.060 * voltage_stability_score - 0.010 * battery_state_pct
        - pe * 0.10 - ie * 0.08 + rng.normal(0, 0.45, n)
    )
    fault_risk = np.where(fault_score >= np.quantile(fault_score, 0.62), 'high_risk', 'normal')
    reading_date = [
        (pd.Timestamp('2025-02-01') + pd.Timedelta(days=int(d))).strftime('%Y-%m-%d')
        for d in rng.integers(0, 390, n)
    ]

    region = region.astype(object)
    inverter_model = inverter_model.astype(object)
    mask = rng.random(n) < 0.08
    region[mask] = [f' {value.upper()} ' for value in region[mask]]
    mask = rng.random(n) < 0.08
    inverter_model[mask] = [value.lower() for value in inverter_model[mask]]

    df_out = pd.DataFrame({
        'site_id': rng.integers(20000000, 99999999, n),
        'region': region,
        'panel_type': panel_type,
        'inverter_model': inverter_model,
        'measurement_month': measurement_month,
        'system_capacity_kw': np.round(system_capacity_kw, 1),
        'solar_irradiance_kwh_m2': np.round(solar_irradiance, 2),
        'cloud_cover_pct': np.round(cloud_cover_pct, 1),
        'ambient_temp_c': np.round(ambient_temp_c, 1),
        'panel_temp_c': np.round(panel_temp_c, 1),
        'humidity_pct': np.round(humidity_pct, 1),
        'wind_speed_m_s': np.round(wind_speed_m_s, 2),
        'dust_index': np.round(dust_index, 2),
        'panel_age_months': panel_age_months,
        'days_since_maintenance': np.round(days_since_maintenance, 0).astype(int),
        'inverter_load_pct': np.round(inverter_load_pct, 1),
        'voltage_stability_score': np.round(voltage_stability_score, 1),
        'battery_state_pct': np.round(battery_state_pct, 1),
        'cleaning_cost_usd': np.round(cleaning_cost_usd, 1),
        'reading_date': reading_date,
        'fault_risk': fault_risk,
        'daily_energy_kwh': np.round(daily_energy_kwh, 1)
    })

    for col, frac in [
        ('solar_irradiance_kwh_m2', 0.05),
        ('cloud_cover_pct', 0.07),
        ('dust_index', 0.06),
        ('voltage_stability_score', 0.04),
        ('panel_type', 0.03),
    ]:
        df_out.loc[rng.random(n) < frac, col] = np.nan

    idx = rng.choice(n, max(6, n // 50), replace=False)
    df_out.loc[idx, 'dust_index'] = (df_out.loc[idx, 'dust_index'] * rng.uniform(1.7, 2.5, len(idx))).round(2)
    idx = rng.choice(n, max(6, n // 55), replace=False)
    df_out.loc[idx, 'panel_temp_c'] = (df_out.loc[idx, 'panel_temp_c'] + rng.uniform(8, 14, len(idx))).round(1)
    return df_out

# -- Student helper toolbox -------------------------------------------------
def quick_audit(data):
    """Compact audit table: dtype, missingness, uniqueness, and examples."""
    return pd.DataFrame({
        'dtype': data.dtypes.astype(str),
        'missing': data.isna().sum(),
        'missing_pct': (data.isna().mean() * 100).round(2),
        'n_unique': data.nunique(dropna=True),
        'sample_values': data.apply(lambda s: list(s.dropna().astype(str).unique()[:3]))
    }).sort_values(['missing_pct', 'n_unique'], ascending=[False, True])


def split_feature_groups(data, targets=(), id_like=()):
    feature_cols = [c for c in data.columns if c not in set(targets) | set(id_like)]
    numeric = data[feature_cols].select_dtypes(include=np.number).columns.tolist()
    categorical = data[feature_cols].select_dtypes(exclude=np.number).columns.tolist()
    date_like = [c for c in feature_cols if 'date' in c.lower() or c.lower().endswith('_dt')]
    return {'numeric': numeric, 'categorical': categorical, 'date_like': date_like}


def clean_category_text(data, columns=None):
    out = data.copy()
    if columns is None:
        columns = out.select_dtypes(include=['object', 'string', 'category']).columns
    for col in columns:
        cleaned = (
            out[col].astype('string')
                    .str.strip()
                    .str.lower()
                    .str.replace(r'\s+', ' ', regex=True)
        )
        out[col] = cleaned.to_numpy(dtype=object, na_value=np.nan)
    return out


def add_date_parts(data, columns, drop_original=True):
    out = data.copy()
    for col in columns:
        dt = pd.to_datetime(out[col], errors='coerce')
        out[f'{col}_month'] = dt.dt.month
        out[f'{col}_dayofweek'] = dt.dt.dayofweek
        out[f'{col}_quarter'] = dt.dt.quarter
        if drop_original:
            out = out.drop(columns=[col])
    return out


def cap_outliers_iqr(data, columns, factor=1.5):
    out = data.copy()
    caps = {}
    for col in columns:
        q1 = out[col].quantile(0.25)
        q3 = out[col].quantile(0.75)
        iqr = q3 - q1
        lower = q1 - factor * iqr
        upper = q3 + factor * iqr
        out[col] = out[col].clip(lower, upper)
        caps[col] = (lower, upper)
    return out, caps


def build_preprocessor(numeric_features, categorical_features, scaler='standard'):
    scaler_map = {
        'standard': StandardScaler(),
        'robust': RobustScaler(),
        'minmax': MinMaxScaler()
    }
    numeric_pipe = Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', scaler_map.get(scaler, StandardScaler()))
    ])
    categorical_pipe = Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])
    return ColumnTransformer([
        ('num', numeric_pipe, numeric_features),
        ('cat', categorical_pipe, categorical_features)
    ])


def cv_classification_summary(models, X, y, cv=5):
    splitter = StratifiedKFold(n_splits=cv, shuffle=True, random_state=42)
    rows = []
    for name, model in models.items():
        scores = cross_validate(
            model, X, y, cv=splitter,
            scoring={'accuracy': 'accuracy', 'f1_macro': 'f1_macro'},
            return_train_score=True
        )
        rows.append({
            'model': name,
            'train_accuracy': scores['train_accuracy'].mean(),
            'cv_accuracy': scores['test_accuracy'].mean(),
            'cv_f1_macro': scores['test_f1_macro'].mean()
        })
    return pd.DataFrame(rows).sort_values('cv_accuracy', ascending=False)


def cv_regression_summary(models, X, y, cv=5):
    splitter = KFold(n_splits=cv, shuffle=True, random_state=42)
    rows = []
    for name, model in models.items():
        scores = cross_validate(
            model, X, y, cv=splitter,
            scoring={'r2': 'r2', 'neg_mae': 'neg_mean_absolute_error'},
            return_train_score=True
        )
        rows.append({
            'model': name,
            'train_r2': scores['train_r2'].mean(),
            'cv_r2': scores['test_r2'].mean(),
            'cv_mae': -scores['test_neg_mae'].mean()
        })
    return pd.DataFrame(rows).sort_values('cv_r2', ascending=False)


def _install_focus_guard(fingerprint):
    # Best effort only: works in classic Notebook; JupyterLab/browser settings may block JS.
    js = f"""
    (function() {{
        const guardKey = 'smart_exam_focus_guard_{fingerprint}';
        try {{ sessionStorage.removeItem(guardKey); }} catch(e) {{}}
        if (window.__smartExamGuardInstalledFor === guardKey) {{ return; }}
        window.__smartExamGuardInstalledFor = guardKey;

        function trigger(reason) {{
            try {{
                if (sessionStorage.getItem(guardKey) === 'triggered') {{ return; }}
                sessionStorage.setItem(guardKey, 'triggered');
            }} catch(e) {{}}
            const msg = 'Notebook focus was lost (' + reason + '). Exam outputs will be cleared and the kernel will restart.';
            try {{ alert(msg); }} catch(e) {{}}
            try {{
                if (window.Jupyter && Jupyter.notebook) {{
                    Jupyter.notebook.clear_all_output();
                    Jupyter.notebook.save_notebook();
                    Jupyter.notebook.kernel.restart();
                    return;
                }}
            }} catch(e) {{}}
            try {{
                document.querySelectorAll('.output, .output_wrapper, .jp-Cell-outputArea').forEach(function(el) {{
                    el.innerHTML = '';
                }});
            }} catch(e) {{}}
            try {{ location.reload(); }} catch(e) {{}}
        }}

        setTimeout(function() {{
            document.addEventListener('visibilitychange', function() {{
                if (document.hidden) {{ trigger('tab hidden'); }}
            }});
            window.addEventListener('blur', function() {{
                setTimeout(function() {{
                    if (!document.hasFocus()) {{ trigger('window blur'); }}
                }}, 900);
            }});
        }}, 3000);
    }})();
    """
    display(Javascript(js))

# ── Main unlock sequence ─────────────────────────────────────────
print('=' * 60)
print('  🔓  SMART EXAM UNLOCK  ·  IN221  ·  Spring 2026')
print('=' * 60)

_reg = input('\n📋  Enter your registration number: ').strip()
if not (_reg.isdigit() and 6 <= len(_reg) <= 12):
    display(HTML("<div style='color:#e94560;font-weight:bold;font-size:15px'>"
                 "❌  Invalid registration number — use digits only.<br>"
                 "Re-run this cell and try again.</div>"))
    raise ValueError('Invalid registration number')

_key = input('🔑  Enter the exam key given by your instructor: ').strip()

_check = _unlock(_VERIFY_TOKEN, _key)
if _check != 'ELPIS_EXAM_2026_OK':
    display(HTML("<div style='color:#e94560;font-weight:bold;font-size:15px'>"
                 "❌  Incorrect exam key. Ask your instructor and re-run this cell.</div>"))
    raise ValueError('Wrong exam key')

_QUESTIONS = json.loads(_unlock(_QUESTION_PAYLOAD, _key))
_seed = int(hashlib.sha256(f'{_key}|{_reg}'.encode()).hexdigest(), 16) % (2**32)
df = _make_dataset(_seed)
_fp = hashlib.sha256(f'{_key}|{_reg}'.encode()).hexdigest()[:12].upper()

display(HTML(f"""
<div style='background:#0d2818;color:#c9d1d9;border-radius:12px;padding:22px;font-family:monospace;border:1px solid #238636;'>
  <h3 style='color:#3fb950;margin-top:0'>✅  Exam Unlocked Successfully</h3>
  <table style='width:100%;border-collapse:collapse'>
    <tr><td style='color:#8b949e;padding:3px 10px'>📋 Registration</td><td style='color:#fff;padding:3px 10px'><strong>{_reg}</strong></td></tr>
    <tr><td style='color:#8b949e;padding:3px 10px'>🔑 Fingerprint</td><td style='color:#fff;padding:3px 10px'><strong>{_fp}</strong><span style='color:#484f58;font-size:11px'> (show to instructor if asked)</span></td></tr>
    <tr><td style='color:#8b949e;padding:3px 10px'>📊 Dataset</td><td style='color:#fff;padding:3px 10px'><strong>{df.shape[0]} rows × {df.shape[1]} columns</strong></td></tr>
    <tr><td style='color:#8b949e;padding:3px 10px'>🎓 Tasks</td><td style='color:#fff;padding:3px 10px'><strong>8 tasks · 100 pts total (+2 bonus) · Good luck!</strong></td></tr>
    <tr><td style='color:#8b949e;padding:3px 10px'>🛡️ Focus Guard</td><td style='color:#fff;padding:3px 10px'><strong>Best-effort browser guard installed</strong></td></tr>
  </table>
</div>
"""))

_install_focus_guard(_fp)

print('\n📊 First 5 rows of YOUR unique dataset:')
display(df.head())
```

```python
# -- Student helper toolbox -------------------------------------------------
def quick_audit(data):
    """Compact audit table: dtype, missingness, uniqueness, and examples."""
    return pd.DataFrame({
        'dtype': data.dtypes.astype(str),
        'missing': data.isna().sum(),
        'missing_pct': (data.isna().mean() * 100).round(2),
        'n_unique': data.nunique(dropna=True),
        'sample_values': data.apply(lambda s: list(s.dropna().astype(str).unique()[:3]))
    }).sort_values(['missing_pct', 'n_unique'], ascending=[False, True])


def split_feature_groups(data, targets=(), id_like=()):
    feature_cols = [c for c in data.columns if c not in set(targets) | set(id_like)]
    numeric = data[feature_cols].select_dtypes(include=np.number).columns.tolist()
    categorical = data[feature_cols].select_dtypes(exclude=np.number).columns.tolist()
    date_like = [c for c in feature_cols if 'date' in c.lower() or c.lower().endswith('_dt')]
    return {'numeric': numeric, 'categorical': categorical, 'date_like': date_like}


def clean_category_text(data, columns=None):
    out = data.copy()
    if columns is None:
        columns = out.select_dtypes(include=['object', 'string', 'category']).columns
    for col in columns:
        cleaned = (
            out[col].astype('string')
                    .str.strip()
                    .str.lower()
                    .str.replace(r'\s+', ' ', regex=True)
        )
        out[col] = cleaned.to_numpy(dtype=object, na_value=np.nan)
    return out


def add_date_parts(data, columns, drop_original=True):
    out = data.copy()
    for col in columns:
        dt = pd.to_datetime(out[col], errors='coerce')
        out[f'{col}_month'] = dt.dt.month
        out[f'{col}_dayofweek'] = dt.dt.dayofweek
        out[f'{col}_quarter'] = dt.dt.quarter
        if drop_original:
            out = out.drop(columns=[col])
    return out


def cap_outliers_iqr(data, columns, factor=1.5):
    out = data.copy()
    caps = {}
    for col in columns:
        q1 = out[col].quantile(0.25)
        q3 = out[col].quantile(0.75)
        iqr = q3 - q1
        lower = q1 - factor * iqr
        upper = q3 + factor * iqr
        out[col] = out[col].clip(lower, upper)
        caps[col] = (lower, upper)
    return out, caps


def build_preprocessor(numeric_features, categorical_features, scaler='standard'):
    scaler_map = {
        'standard': StandardScaler(),
        'robust': RobustScaler(),
        'minmax': MinMaxScaler()
    }
    numeric_pipe = Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', scaler_map.get(scaler, StandardScaler()))
    ])
    categorical_pipe = Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])
    return ColumnTransformer([
        ('num', numeric_pipe, numeric_features),
        ('cat', categorical_pipe, categorical_features)
    ])


def cv_classification_summary(models, X, y, cv=5):
    splitter = StratifiedKFold(n_splits=cv, shuffle=True, random_state=42)
    rows = []
    for name, model in models.items():
        scores = cross_validate(
            model, X, y, cv=splitter,
            scoring={'accuracy': 'accuracy', 'f1_macro': 'f1_macro'},
            return_train_score=True
        )
        rows.append({
            'model': name,
            'train_accuracy': scores['train_accuracy'].mean(),
            'cv_accuracy': scores['test_accuracy'].mean(),
            'cv_f1_macro': scores['test_f1_macro'].mean()
        })
    return pd.DataFrame(rows).sort_values('cv_accuracy', ascending=False)


def cv_regression_summary(models, X, y, cv=5):
    splitter = KFold(n_splits=cv, shuffle=True, random_state=42)
    rows = []
    for name, model in models.items():
        scores = cross_validate(
            model, X, y, cv=splitter,
            scoring={'r2': 'r2', 'neg_mae': 'neg_mean_absolute_error'},
            return_train_score=True
        )
        rows.append({
            'model': name,
            'train_r2': scores['train_r2'].mean(),
            'cv_r2': scores['test_r2'].mean(),
            'cv_mae': -scores['test_neg_mae'].mean()
        })
    return pd.DataFrame(rows).sort_values('cv_r2', ascending=False)
```
