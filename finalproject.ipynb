```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
  

glass_identification = fetch_ucirepo(id=42)
df = glass_identification['data']['original']# or glass_identification.original
# or df = glass_identification['data'] if you only want feature columns
df.head()
```

The first thing we can see is the column Id_number which is not of any importance to us and may interfere in numerical calculations, for that reason we will drop and check for duplicates/nulls in the dataset

```python
df.drop('Id_number', axis=1, inplace=True)
print(df.isna().sum())
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
print(df.duplicated().sum())
```
